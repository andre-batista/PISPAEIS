## Shape error - $\zeta_S$

```python
error = np.zeros((N, 2))
for n in range(N):
    error[n,0] = casestudies[n].results[0].zeta_s[-1]
    error[n,1] = casestudies[n].results[1].zeta_s[-1]
plt.plot(chi, error, 'o--', linewidth=5, markersize=15, label=['OSM', 'SOM'])
plt.xlabel(r'$\chi$')
plt.ylabel(r'$\zeta_s$ [%]')
plt.legend(loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

## DNL

```python
dnl = np.zeros(N)
for n in range(N):
    dnl[n] = casestudies[n].test.dnl
plt.plot(chi, dnl, 'o--', linewidth=5, markersize=15)
plt.xlabel(r'$\chi$')
plt.ylabel('DNL')
plt.grid()
plt.tight_layout()
plt.show()
```
