```python
import sys
sys.path.insert(1, '../../../eispy2d/library/')
import casestudy as cst
import numpy as np
from matplotlib import pyplot as plt
import result as rst
```

```python
N = 5
fig, ax, _ = rst.get_figure(nsubplots=N)
casestudies = [cst.CaseStudy(import_filename='vary%d.cst' % n,
                             import_filepath='./data/') for n in range(N)]
chi = np.array([.25, .5, .75, 1., 1.5])

for n in range(N):
    casestudies[n].reconstruction(axis=[ax[n]], title=r'$\chi$ = %.2f' % chi[n],
                                  fontsize=15)

plt.tight_layout()
plt.show()
```

```python
error = np.zeros(N)
for n in range(N):
    error[n] = casestudies[n].results.zeta_s[-1]
plt.plot(chi, error, 'o--')
plt.xlabel(r'$\chi$')
plt.ylabel(r'$\zeta_s$ [%]')
plt.grid()
plt.show()
```

```python
dnl = np.zeros(N)
for n in range(N):
    dnl[n] = casestudies[n].test.dnl
plt.plot(chi, dnl, 'o--')
plt.xlabel(r'$\chi$')
plt.ylabel('DNL')
plt.grid()
plt.show()
```
