```python
import numpy as np
from matplotlib import pyplot as plt

def rotate_axis(theta, x, y):
    """Summarize the method."""
    T = np.array([[np.cos(theta), np.sin(theta)],
                  [-np.sin(theta), np.cos(theta)]])
    r = np.vstack((x.reshape(-1), y.reshape(-1)))
    rp = T@r
    xp, yp = np.vsplit(rp, 2)
    xp = np.reshape(np.squeeze(xp), x.shape)
    yp = np.reshape(np.squeeze(yp), y.shape)
    return xp, yp
```

```python
color = ['r', 'g', 'b', 'c', 'm', 'y', 'k', 'gray']

center = [0.2, 0.2]
Lx = Ly = 4.

a = .4
circle = np.array([a*np.cos(np.linspace(0, 2*np.pi, 100)), 
                   a*np.sin(np.linspace(0, 2*np.pi, 100))])
rho = 0.8
theta = np.array([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi, 5*np.pi/4, 3*np.pi/2, 7*np.pi/4])
ant = np.array([rho*np.cos(theta), rho*np.sin(theta)])
ant[0] -= center[0]
ant[1] -= center[1]

thetas = 0# np.pi/10

vec = np.array([-1.5, -0.2])
vec = rotate_axis(thetas, vec[0], vec[1])
ang = np.array([1., 0.])
ang = rotate_axis(thetas, ang[0], ang[1])

plt.plot(circle[0], circle[1], label='circle')
for n in range(theta.size):
    plt.plot(ant[0, n], ant[1, n], 'o', color=color[n])
    plt.text(ant[0, n], ant[1, n], str(n), color=color[n])
# plt.plot(ant[0], ant[1], 'o', color=color, label='antenna')
plt.quiver(vec[0], vec[1], ang[0], ang[1], color='b', label='vector')
plt.axis('equal')
plt.xlim(-Lx/2, Lx/2)
plt.ylim(-Ly/2, Ly/2)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
