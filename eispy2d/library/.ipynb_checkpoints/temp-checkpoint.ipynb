```python
pso = ea.ParticleSwarmOptimization()
print(pso.w, pso.c1, pso.c2)
if pso.pbest is not None or pso.gbest is not None or pso.va is not None or pso.vb:
    print('Error')
```

```python
config = cfg.Configuration(name='std', wavelength=1.,perfect_dielectric=True, image_size=[4., 4.], number_measurements=25, number_sources=25)
```

```python
disc = ea.CollocationMethod(config,
                            (5, 5),
                            (5, 5, 25),
                            ea.Bilinear(config.Lx),
                            ea.Bilinear(config.Lx),
                            phi_trial=ea.Bilinear(config.Lx))
```

```python
# fx = ea.DataResiduals()
fx = ea.Rastrigin()
```

```python
alg = ea.EvolutionaryAlgorithm(config, pso, disc, 100, 100, 100000000, 1, fx)
```

```python
epsilon_r, sigma = exp.draw_square(1., 4., 4., (70, 70), object_relative_permittivity=1.25, center=[1.3, 1.3])
instance = ipt.InputData(name='basic', configuration_filename='std', resolution=(40, 40), relative_permittivity_map=epsilon_r, conductivity_map=sigma)
```

```python
fwr = mom.MoM_CG_FFT(config)
_ = fwr.solve(instance, SAVE_INTERN_FIELD=False)
```

```python
alg.max_iterations = 1000
result = alg.solve(instance)
```
